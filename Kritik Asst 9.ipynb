```python
import sympy as sp
import nump as np
import matplotlib.pyplot as plt

#question 4a
x,y=sp.symbols ('x y')
f=sp.exp(x)*sp.sin(y)+y**3

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("∂f/∂x:", df_dx)
print("∂f/∂y:", df_dy)

#question 4b
g = x**2 * y + x * y**2

grad_g = [sp.diff(g, x), sp.diff(g, y)]
magnitude = sp.sqrt(grad_g[0].subs({x: 1, y: -1})**2 + grad_g[1].subs({x: 1, y: -1})**2)

print("Gradient of g(x, y):", grad_g)
print("Magnitude at (1, -1):", magnitude)

#question 4c
h = sp.ln(x**2 + y**2)

h_xx = sp.diff(h, x, x)
h_yy = sp.diff(h, y, y)
h_xy = sp.diff(h, x, y)

print("∂²h/∂x²:", h_xx)
print("∂²h/∂y²:", h_yy)
print("∂²h/∂x∂y:", h_xy)

#question 4d
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

#question 4e
k = sp.sin(x) * sp.cos(y)
k_func = sp.lambdify((x, y), k, 'numpy')

Z_k = k_func(X, Y)

plt.contourf(X, Y, Z_k, levels=50, cmap='plasma')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = sin(x) * cos(y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
