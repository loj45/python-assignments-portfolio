```python
import numpy as np
from mpl_toolkits import mplot3d #for 3D plots
import matplotlib.pyplot as plt #usual matplotlib
%matplotlib widget 

def gradient_descent (x0, y0, f, grad_f, alpha, num_iterations):
    """
    Performs gradient descent to find a local minimum of a function.

    Parameters:
    x0, y0: Initial coordinates (starting point)
    f: Function of two variables (to minimize)
    grad_f: Function that returns the gradient (partial derivatives)
    alpha: Learning rate (step size)
    num_iterations: Maximum number of iterations

    Returns:
    (x, y): Final coordinates after gradient descent.
    """
    
    x, y = x0, y0  # Initialize with starting point

    for i in range(num_iterations): #obtain the gradient of f at (x,y)
        grad_x, grad_y = grad_f(x, y)  # Compute gradient at (x, y)
        x -= alpha * grad_x  # Update x by moving in the opposite direction of gradient
        y -= alpha * grad_y  # Update y similarly
        
        if i % 1000 == 0:
            print(f"Iteration {i}: x = {x:.5f}, y = {y:.5f}, f(x,y) = {f(x,y):.5f}")
            
    return x, y

def fun_1(x, y):
    return x**2 + y**2

def grad_f_1(x, y):
    return (2*x, 2*y)

# Function f2(x, y) and its gradient
def fun_2(x, y):
    return 1 - np.exp(-x**2 - (y - 2)**2) - 2*np.exp(-x**2 - (y + 2)**2)

def grad_f_2(x, y):
    grad_x = 2*x * (np.exp(-x**2 - (y - 2)**2) + 2*np.exp(-x**2 - (y + 2)**2))
    grad_y = 2*(y-2)*np.exp(-x**2 - (y-2)**2) + 4*(y+2)*np.exp(-x**2 - (y+2)**2)
    return (grad_x, grad_y)

print("Running gradient descent for f1(x,y) starting at (0.1, 0.1):")
x1, y1 = gradient_descent(0.1, 0.1, fun_1, grad_f_1, 0.1, 10)
print(f"Final result: x = {x1:.5f}, y = {y1:.5f}\n")

print("Running gradient descent for f1(x,y) starting at (-1, 1):")
x2, y2 = gradient_descent(-1, 1, fun_1, grad_f_1, 0.01, 100)
print(f"Final result: x = {x2:.5f}, y = {y2:.5f}\n")

# Run gradient descent on f2(x, y)
print("Running gradient descent for f2(x,y) starting at (0,1):")
x3, y3 = gradient_descent(0, 1, fun_2, grad_f_2, 0.01, 10000)
print(f"Final result: x = {x3:.5f}, y = {y3:.5f}\n")

print("Running gradient descent for f2(x,y) starting at (0,-1):")
x4, y4 = gradient_descent(0, -1, fun_2, grad_f_2, 0.01, 10000)
print(f"Final result: x = {x4:.5f}, y = {y4:.5f}\n")

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(X, Y)
z = fun_2(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
plt.title("3D plot of kritik 8 function2(x, y)")
plt.show()
```
