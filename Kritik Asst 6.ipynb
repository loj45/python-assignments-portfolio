```python
import numpy as np
import matplotlib.pyplot as plt

def normal_density(mean, variance, x):
    formula = 1/np.sqrt(2*np.pi*variance**2)*np.exp(-(x-mean)**2/(2*variance**2))
    return formula

def plot(mean, variance):
    nd_x=[]
    nd_y=[]

    for i in range(0, 25):
        nd_x.append(i)
        f=normal_density(mean,variance,i)
        nd_y.append(f)

    return nd_x, nd_y

plt.title("Plot of Normal Density with Different Means and Variances")

nd_x, nd_y = plot(2,3)
plt.scatter(nd_x,nd_y)
plt.plot(nd_x,nd_y)

nd_x1, nd_y1 = plot(14,7)
plt.scatter(nd_x1,nd_y1)
plt.plot(nd_x1,nd_y1)

nd_x2, nd_y2 = plot(2,11)
plt.scatter(nd_x2,nd_y2)
plt.plot(nd_x2,nd_y2)

plt.grid()
plt.show()
```

```python
def form(mean, variance, x):
    f = 1/np.sqrt(2*np.pi*variance**2)*np.exp(-(x-mean)**2/(2*variance**2))
    return f

def integration(mean, variance, a, b):
    
    x = np.linspace(a, b, 100)  
    y = form(mean, variance, x)  
    dx = (b - a) / (99)
    integral = 0

    for i in range(99):
        integral += (y[i] + y[i + 1]) * dx / 2
        
    return integral

print(integration(171, 7.1, 162, 190))
```

```python
def form(mean, variance, x):
    f = 1/np.sqrt(2*np.pi*variance**2)*np.exp(-(x-mean)**2/(2*variance**2))
    return f

def dosage (mean, variance, a, b):
    def dosage_function(x):
        return 2.38*(x**2)
    
    x = np.linspace(a, b, 100)  
    normal_distr=form(mean, variance,x)
    dosage=dosage_function(x)
    
    dx = (b - a) / (99)
    integral = 0

    for i in range(99):
        integral += (normal_distr[i] + normal_distr[i+1]) * (dosage[i] + dosage[i+1]) * dx/4
        
    return integral

print(dosage(171, 7.1, 162, 190))
```
